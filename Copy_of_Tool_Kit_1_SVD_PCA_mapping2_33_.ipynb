```python
from scipy.stats import norm
import numpy as np
from scipy.linalg import eig
from numpy import cov
from numpy import argsort
from numpy import var
#Sample 1000 points from a Gaussian distribution with mean 0 and standard deviation 1
y1 = norm.rvs(size=(1,1000))
#Sample 1000 points from a Gaussian distribution with mean 0 and standard deviation 2
y2 = 2*norm.rvs(size=(1,1000))
#Sample 1000 points from a Gaussian distribution with mean 0 and standard deviation 5
y3 = 5*norm.rvs(size=(1,1000))
print('Maximum variance across y variables: '+ str(max(var(y1),var(y2),var(y3))))
#Generate the x variables from the y variables
x1 = y1
x2 = (y2+y3)/np.sqrt(2)
x3 = (y2-y3)/np.sqrt(2)
# x1, x2, x3 contain the three dimesnions of data matrix X
X= np.concatenate((x1,x2,x3), axis=0)
print('Maximum variance across x variables: '+ str(max(var(x1),var(x2),var(x3))))
X
print(X.shape)
#This is r* d matrix 
Xbar=X.T
print(X)
```
